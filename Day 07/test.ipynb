```python
cap = cv2.VideoCapture('images/11_output.avi')

frame_id = 0
while(cap.isOpened()):
    ret, frame = cap.read()
    if(ret == True):
        frame_id += 1
        cv2.imshow('Label',frame)
        cv2.imwrite(f'imgs/output_{frame_id}.png',frame)
    else:
        break
```
