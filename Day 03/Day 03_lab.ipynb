#Q5- Accept the marked price from the user and calculate the Net amount as(Marked Price â€“ Discount) to pay according to following criteria:

```python
marked_price = float(input("Enter the price: "))
if marked_price > 10000:
  print(f'You should pay {round(marked_price - (marked_price*0.2),2)}')
elif (marked_price <= 10000) and (marked_price > 7000):
  print(f'You should pay {round(marked_price - (marked_price*0.15),2)}')
else:
  print(f'You should pay {round(marked_price - (marked_price*0.1),2)}')
```

# Q8 - Write a program that keep on accepting number from the user until user enters Zero. Display the sum and average of all the numbers.

```python
sum = 0
count = 0
num = int(input())
while num != 0:
  sum += num
  count += 1
  num = int(input())
print(f"Sum is: {sum}")
print(f"Sum is: {round((sum/count),2)}")
```

#Q10- Make Method to Return the Maximum and Minimum item from the given list. lst=[100,30,22,40,60,20,10,300]

```python
def min_max(lst):
  lst.sort()
  return lst[0],lst[-1] 
 
lst = [100,30,22,40,60,20,10,300]
print(f'Max is :{min_max(lst)[1]}\nMin is :{min_max(lst)[0]}')
```

# HackerRank Leap Year

```python
def is_leap(year):
    return (year % 4 == 0) and (year % 100 != 0) or (year % 400 == 0)
year = int(input())
print(is_leap(year))
```
